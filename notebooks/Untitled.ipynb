```python
from glob import glob
import numpy as np
import pandas as pd
import torch
from torch.utils.data import Dataset, DataLoader

IMAGE_SHAPE = (512, 512)


transformer = albu.Compose([
    albu.ShiftScaleRotate(
        shift_limit=0.01,
        scale_limit=(-0.2, 0.1),
        rotate_limit=179,
        border_mode=cv2.BORDER_CONSTANT,
        value=255,
        p=0.95),
])


class ImageReader:
    def load_images(self, images_dir: str = 'images'):
        df = pd.DataFrame(columns=['words', 'image_array'])

        for path in glob(f'{images_dir}/*.jpg'):
            new_row = {
                'words': self.get_words(path),
                'image_array': self.get_image(path),
            }
            df = df.append(new_row, ignore_index=True)

        return df

    @staticmethod
    def get_words(path: str):
        return ' '.join(re.findall('[A-Z][^A-Z]*', path.split('/')[-1].split('.')[0]))

    @staticmethod
    def get_image(path: str):
        return cv2.resize(cv2.cvtColor(cv2.imread(path), cv2.COLOR_BGR2GRAY), IMAGE_SHAPE)


class LogogramDataset(Dataset):
    def __init__(self, df, transformer, add_noise=False):
        self.df = df
        self.add_noise = add_noise
        self.transformer = transformer
        self.noise_transformer = albu.Cutout(
            num_holes=100, max_h_size=7, max_w_size=7, p=.75
        )

    def __getitem__(self, idx):
        sentence = self.df['words'].iloc[idx]
        sentence = self._process_words(sentence)
        output_img = self.df['image_array'].iloc[idx]
        output_img = self.transformer(image=output_img)["image"]
        
        if self.add_noise:
            input_img = self.noise_transformer(image=output_img)["image"]
        else:
            input_img = output_img
        
        input_img = torch.from_numpy(np.expand_dims(input_img, axis=0)) / 255
        output_img = torch.from_numpy(np.expand_dims(output_img, axis=0)) / 255

        return sentence, input_img, output_img

    def __len__(self):
        return len(self.df)
    
    @staticmethod
    def _process_words(sentence):
        sentence = sentence.replace('_', '').split()
        idx = np.random.randint(len(sentence))

        return ' '.join(sentence[idx:] + sentence[:idx])
```

```python
import torch.nn as nn
import torch.nn.functional as F

from transformers import AutoTokenizer, AutoModel


class LogoTranslator(nn.Module):
    def __init__(self, pretrain_model: str = 'albert-base-v2'):
        super().__init__()
        self.tokenizer = AutoTokenizer.from_pretrained(pretrain_model)
        self.albert = AutoModel.from_pretrained(pretrain_model)
        self.lstm = nn.LSTM(
            input_size=768, hidden_size=32 * 12 * 12, num_layers=1, batch_first=True
        )
        self.decoder = nn.Sequential(
            nn.ConvTranspose2d(32, 16, 3, stride=2, dilation=2),  # 16 * 27 * 27
            nn.BatchNorm2d(16),
            nn.GELU(),
            nn.ConvTranspose2d(16, 8, 3, stride=3, dilation=2),  # 8 * 83 * 83
            nn.BatchNorm2d(8),
            nn.GELU(),
            nn.ConvTranspose2d(8, 4, 2, stride=3, dilation=2),  # 4 * 249 * 249
            nn.BatchNorm2d(4),
            nn.GELU(),
            nn.ConvTranspose2d(4, 1, 4, stride=2, dilation=5),  # 1 * 512 * 512
            nn.BatchNorm2d(1),
            nn.GELU(),
        )
        self.deblur = nn.Sequential(
            nn.Conv2d(1, 32, 9, padding=2, padding_mode='replicate'),
            nn.GELU(),
            nn.Conv2d(32, 4, 1, padding=2, padding_mode='replicate'),
            nn.GELU(),
            nn.Conv2d(4, 1, 5, padding=2, padding_mode='replicate'),
        )

    def forward(self, x):
        x = self.tokenizer(x, padding=True, return_tensors='pt')
        x = self.albert(**x).last_hidden_state
        x, _ = self.lstm(x)
        x = x[:, -1, :].view(-1, 32, 12, 12)
        x = self.decoder(x)
        x = self.deblur(x)

        return x
```

```python
from torch.optim import Adam
from lovasz_losses import lovasz_hinge

N_EPOCHS = 20
BATCH_SIZE = 16
LEARNING_RATE = 0.001


image_reader = ImageReader()
dataset = LogogramDataset(image_reader.load_images(), transformer, add_noise=False)
data_loader = DataLoader(
    dataset, batch_size=BATCH_SIZE, shuffle=False, num_workers=0
)

train_loss = []
model = LogoTranslator()
criterion = lovasz_hinge
# criterion = nn.BCELoss()
optimizer = Adam(model.parameters(), lr=LEARNING_RATE)

for epoch in range(1, N_EPOCHS + 1):
    epoch_loss = 0
    for data in data_loader:
        optimizer.zero_grad()

        words, _, output_img = data
        outputs = model(list(words))

        # calculate the loss
        loss = criterion(outputs, output_img)
        loss.backward()
        optimizer.step()
        train_loss.append(loss.item())
        epoch_loss += loss.item() * len(words)

    # print avg training statistics
    if (epoch % 3 == 1) or (epoch == N_EPOCHS):
        epoch_loss = epoch_loss / len(data_loader)
        print('Epoch: {} \tTraining Loss: {:.6f}'.format(epoch, epoch_loss))
```

```python
from stack_gan.generator import GeneratorNet
from stack_gan.discriminator import DiscriminatorResN
import math
from torch.optim import Adam
from torch import nn


TEXT_DIM = 512
ENBEDDING_DIM = 512
GEN_CHANNELS = 128
GEN_LR = 0.001
DIS_CHANNELS = 128
DIS_LR = 0.001


model = GANTrainer(
    models={
        "generator": GeneratorNet(TEXT_DIM, ENBEDDING_DIM, NOISE_DIM, GEN_CHANNELS, IMAGE_SIZE),
        "discriminator": [
            DiscriminatorResN(DIS_CHANNELS, TEXT_DIM, ENBEDDING_DIM, 2**r)
            for r in range(6, int(math.log2(IMAGE_SIZE)))
        ],
    },
    optimizers={"generator": Adam, "discriminator": Adam},
    criterions={"generator": nn.BCELoss(), "discriminator": nn.BCELoss()},
    optimizer_params={"gen_lr": GEN_LR, "dis_lr": DIS_LR},
)
```
